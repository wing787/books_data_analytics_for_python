## 4.3.5 pandasのオブジェクトからグラフ描画

```python
import numpy as np
import pandas as pd
import matplotlib.style
import matplotlib.pyplot as plt


# ggplotスタイルを指定
matplotlib.style.use('ggplot')
```

```python
# DataFrameを作成
df = pd.DataFrame({'A': [1, 2, 3], 'B': [3, 1, 2]})
df.plot()  # 折れ線グラフを描画
plt.show()
```

```python
# ランダムに3行2列のデータを生成
rng = np.random.default_rng(123)
df2 = pd.DataFrame(rng.random((3, 2)), columns=['y1', 'y2'])

df2.plot.bar()  # 棒グラフを描画
plt.show()
```

```python
df2.plot.bar(stacked=True)  # 積み上げ棒グラフを描画
plt.show()
```
